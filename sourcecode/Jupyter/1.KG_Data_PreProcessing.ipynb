# Customer Sentiment Analysis using Knowledge Graph

Synopysis

Objective – Create the knowledge graph of customer product feedback to do customer sentiment analysis. Customer Product feedback is present in dataset. That includes customer (reviewer) details, Productname, Product category, Product rating given by the reviewer.
The focus of the research is to derive entity relationship based on the customer sentiment's to showcase data insight. Using any other ML methods unknown data insight discovery is almost impossible. in this research we will leverage Knowldge graph and relative concept to perform this discovery.  

Data Source - For this research we have used e-commerce site-Amazon product review data. Amozon product review is available as public dataset. Data is avialble at Kaggle and other hosted Site in different format as mentioned below :
  
   Data type                Source  
   csv             https://data.world/datafiniti/consumer-reviews-of-amazon-products/workspace/file?filename=Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv

```python
# Import libraries
# Pandas for :
# Warning for :
# Seaborn for :

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import seaborn as sns
```

#### 2. Data Importing

```python
##Importing CSV dataset to Pandas dataframe
pddf=pd.read_csv('Datafiniti_Amazon.csv')
pddf.head(5)
```

```python
#--------------------------------------------------------------
#Size and shape of the dataset - There are 28,000 observations and 25 columns
#--------------------------------------------------------------

print("Shape of data:", pddf.shape)

print("Numerical features count: ", len(pddf.dtypes[pddf.dtypes != "object"].index))

print("Categorical features count: ", len(pddf.dtypes[pddf.dtypes == "object"].index))
```

#### 2. Data Quality checks

```python
#--------------------------------------------------------------
#1. Check for missing/null values - There are no missing values
#--------------------------------------------------------------
pddf.isnull().sum()
```

Above null values feature columns are not included in the processing. These column will be removed in forward steps.

```python
#--------------------------------------------------------------
#2. Check for duplicates
#--------------------------------------------------------------

print("No of unique records: ", pddf['dateAdded'].nunique())
```

```python
#--------------------------------------------------------------
#3. Elimination of unnecessary feature variable and data Cleaning
#--------------------------------------------------------------

# Dropping unnecessary features Ref: as per result of Step 1
pddf = pddf.drop(["id","dateAdded","dateUpdated","asins","imageURLs","keys","manufacturerNumber","reviews.dateSeen","reviews.id","reviews.numHelpful","reviews.sourceURLs","reviews.text","reviews.title","sourceURLs"],axis=1)

#Renaming name feature to product name 
pddf.rename(columns={'name':'ProductName'}, inplace=True)
pddf.rename(columns={'reviews.username':'ReviewerName'}, inplace=True)
pddf.rename(columns={'brand':'Brand'}, inplace=True)
pddf.rename(columns={'categories':'Categories'}, inplace=True)
pddf.rename(columns={'primaryCategories':'PrimaryCategories'}, inplace=True)
pddf.rename(columns={'manufacturer':'Manufacturer'}, inplace=True)
pddf.rename(columns={'reviews.date':'ReviewDate'}, inplace=True)
pddf.rename(columns={'reviews.rating':'ReviewRating'}, inplace=True)
pddf.rename(columns={'reviews.doRecommend':'doRecommend'}, inplace=True)
pddf.rename(columns={'reviews.didPurchase':'didPurchase'}, inplace=True)
pddf.head(5)
```

```python
#--------------------------------------------------------------
# 4. Data Correctness check 
# Feature column to check 
# reviews.rating should be between 1 to 5
#--------------------------------------------------------------
print(pddf.ReviewRating[pddf.ReviewRating<1].count(),
      pddf.ReviewRating[pddf.ReviewRating>5].count())
```

```python
#
# Distribution of feature variables 
#
print("=========== Data Distribution=============")
print(pddf['PrimaryCategories'].value_counts())
print("_"*100)
print(pddf['ProductName'].value_counts())
print("_"*100)
print(pddf['ReviewerName'].value_counts())
print("_"*100)
```

# Saving cleansed file in csv format for further processing

```python
## Treat Missing Value of doRecommend for record with Blank
pddf['doRecommend'].fillna('NotGiven', inplace=True)
## Derive Feature column of doRecommend categoricial variable to numericial variable
pddf['Recommend']=1 
pddf.loc[pddf.doRecommend=='True',"Recommend"]=1
pddf.loc[pddf.doRecommend=='False',"Recommend"]=-1
pddf.loc[pddf.doRecommend=='NotGiven',"Recommend"]=0
## Transform Rating value 1 to 2 to Negative sentiments. 1 is lesser rating then 2 hence in scoring we will score -2 t0 1 and -1 to 2.
pddf.loc[pddf.ReviewRating==1,"NegativeScore"]=-2
pddf.loc[pddf.ReviewRating==2,"NegativeScore"]=-1
## Review rating equal or more than 3 are with good rating hence negative score is zero for such review record.
pddf.loc[pddf.ReviewRating>=3,"NegativeScore"]=0
## Transform Rating value 3 to 5 to Positive sentiments. 3 is lesser rating then 4 hence in scoring we will score 1 for 3, 2 for 4 and 3 for 5.
pddf.loc[pddf.ReviewRating==3,"PositiveScore"]=1
pddf.loc[pddf.ReviewRating==4,"PositiveScore"]=2
pddf.loc[pddf.ReviewRating==5,"PositiveScore"]=3
## Review rating lesser than 3 are with bad rating hence positive score is zero for such review record.
pddf.loc[pddf.ReviewRating<3,"PositiveScore"]=0
pddf
```

```python
pddf.to_csv("amazon_df_processed.csv", index=None)
```

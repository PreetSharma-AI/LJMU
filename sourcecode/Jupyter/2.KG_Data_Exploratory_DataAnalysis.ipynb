### 2. Bivariate Analysis


```python
def bivariate_eda(df, cols, hue, n_cols = 1, figsize = None):

    n_rows = len(cols)/n_cols
    plt.figure(figsize=(5*n_cols,5*n_rows))
    
    if figsize:
        plt.figure(figsize=figsize)
    
    for i in range(0,len(cols)):
        plt.subplot(n_rows+1, n_cols, (i+1))
        ax = plt.gca()
        ax.tick_params(axis='x', labelrotation = 60)
        sns.countplot(x=df[cols[i]],hue=df[hue],palette="Blues")
 
    plt.show()

bivariate_eda(inpt_df,['Brand'], 'ReviewRating', n_cols = 1, figsize = (13,8))
bivariate_eda(inpt_df,['PrimaryCategories'], 'ReviewRating', n_cols = 1, figsize = (13,8))
bivariate_eda(inpt_df,['Manufacturer'], 'ReviewRating', n_cols = 1, figsize = (13,8))
bivariate_eda(inpt_df,['Recommend'], 'ReviewRating', n_cols = 1, figsize = (13,8))

```

```python
import seaborn as sns

sns.set_style("whitegrid")
#tips = sns.load_dataset("tips")
box_plot = sns.boxplot(x="PrimaryCategories",y="ReviewRating",data=inpt_df)

medians = inpt_df.groupby(['ProductName'])['ReviewRating'].median()
vertical_offset = inpt_df['ReviewRating'].median() * 0.05 # offset from median for display

for xtick in box_plot.get_xticks():
    box_plot.text(xtick,medians[xtick] + vertical_offset,medians[xtick], 
            horizontalalignment='center',size='x-small',color='w',weight='semibold')
    box_plot.set_xticklabels(box_plot.get_xticklabels(),rotation=90)
```
